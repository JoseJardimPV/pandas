## Transformação dos dados

```python
# criando uma coluna com o total de todos os stats
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
```

```python
# para vizualizar o df sem a coluna
df.drop(columns=['Total'])
```

```python
# para remover a coluna criada
df.drop(columns=['Total'], inplace=True)
```

```python
df
```

```python
# outra forma de criar a coluna "Total"
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
```

```python
df
```

```python
# reordenando o df
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
```

```python
df
```

# Filtrando os dados

## Filtro 'grass' 'poison' 'hp'

```python
# Filtrando com 'and'(&) e 'or'(|)
df.loc[(df['Type 1'] ==  'Grass') & (df['Type 2'] == 'Poison') & (df['HP']>70)]
```

```python
# criando um df só com os novos as informações anteriores
df_gp = df.loc[(df['Type 1'] ==  'Grass') & (df['Type 2'] == 'Poison') & (df['HP']>70)]
```

```python
df_gp
```

```python
# criando um novo index no novo df, começando do zero (drop p/ remover o atual index)
df_gp = df_gp.reset_index(drop=True)
```

```python
df_gp
```

## Filtro 'mega'

```python
# para filtrar todos que contém 'mega' no nome
df.loc[df['Name'].str.contains('Mega')]
# caso queira filtrar o oposto (não contém 'Mega") adicionar ~
# df.loc[~df['Name'].str.contains('Mega')]
```

## Filtros regex

```python
import re
```

```python
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
# para ignorar a diferença entre minúsculas e maiúsculas pode usar "flags=re.I" da biblioteca regex
df.loc[df['Type 1'].str.contains('fiRe|GraSS', flags=re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

## Mudanças condicionais

```python
# trocando 'fire' por 'fogo'
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Fogo'
```

```python
df
```

```python
# condições podem ser usadas para alterar outras colunas com o mesmo valor ou valores diferentes
df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = ['Teste1', 'Teste2']
```

```python
df
```

## Agregações, estatíticas e análises

```python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
```

```python
# exemplo de uma agregação sem contexto
df.groupby(['Type 1']).sum()
```

```python
df.groupby(['Type 1']).count()
```

```python
# uma forma de ter uma contagem "mais limpa" da quantidade de linhas agregando o tipo 1
df['count'] = 1
df.groupby(['Type 1']).count()['count']
```

```python
df.groupby(['Type 1', 'Type 2']).count()['count']
```
